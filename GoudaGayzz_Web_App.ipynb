```
#Load all figures
import plotly.express as px
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import numpy as np
from dash.dependencies import Input, Output# Load Data

#Dummy graphs for placeholders

df_bar = pd.DataFrame({
    "Fruit": ["Apples", "Oranges", "Bananas", "Apples", "Oranges", "Bananas"],
    "Amount": [4, 1, 2, 2, 4, 5],
    "City": ["SF", "SF", "SF", "Montreal", "Montreal", "Montreal"]
})

fig = px.bar(df_bar, x="Fruit", y="Amount", color="City", barmode="group")

import plotly.graph_objects as go

labels = ['Oxygen','Hydrogen','Carbon_Dioxide','Nitrogen']
values = [4500, 2500, 1053, 500]

# Use `hole` to create a donut-like pie chart
fig_donut = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.6)])

#Fields pie chart

Eng_count1 = data_all_years_remapped['General_STEM_Field'].str.contains("Engineering")
Eng_count = Eng_count1[Eng_count1==True].shape[0]

Sci_count1 = data_all_years_remapped['General_STEM_Field'].str.contains("Science")
Sci_count = Sci_count1[Sci_count1==True].shape[0]

Math_count1 = data_all_years_remapped['General_STEM_Field'].str.contains("Math")
Math_count = Math_count1[Math_count1==True].shape[0]

Tech_count1 = data_all_years_remapped['General_STEM_Field'].str.contains("Technology")
Tech_count = Tech_count1[Tech_count1==True].shape[0]

STEM_Field = pd.DataFrame({'Field': ["Science","Technology","Engineering","Math"], 
                          'Count': [Sci_count,Tech_count,Eng_count,Math_count]})

fig_fields = px.pie(STEM_Field, values='Count', names='Field', 
             title='oSTEM General STEM Field', 
             color_discrete_sequence=px.colors.qualitative.G10)

fig_fields.update_layout(
    margin={"r":300,"t":100,"l":100,"b":75}
)

# Student/Professional Participation
Student = data_all_years_remapped['Ticket'].str.contains("Student/Postdoc")
Student_count = Student[Student==True].shape[0]

Prof = data_all_years_remapped['Ticket'].str.contains("Professional")
Expo = data_all_years_remapped['Ticket'].str.contains("Expo Only")
Staff = data_all_years_remapped['Ticket'].str.contains("oSTEM Staff")


Student_count = Student[Student==True].shape[0]
Professional_count = Prof[Prof==True].shape[0] + Expo[Expo==True].shape[0]
Staff_count = Staff[Staff==True].shape[0]

Student_prof = pd.DataFrame({'Group': ["Students","Professional","Staff"], 
                          'Number': [Student_count,Professional_count,Staff_count]})

fig_student_prof = px.pie(Student_prof, values='Number', names='Group', 
             title='oSTEM Conference Participants', 
             color_discrete_sequence=px.colors.qualitative.G10)

#Major and Departments
df_years = data_all_years_remapped['Major_or_Department'].value_counts()
df_years = pd.DataFrame(df_years)
df_years = df_years[1:20]
df_years['major/dep'] = list(df_years.index)
df_years.columns = ['count','major/dep']

fig_major_dept = px.pie(df_years, values='count', names='major/dep', 
             title='oSTEM Conference Participants Majors and Departments', 
             color_discrete_sequence=px.colors.qualitative.G10)

#Attendee Breakdown

data_subset = data_all_years_remapped[['Ticket','survey_year']]

stu_post = data_all_years_remapped['Ticket'].str.contains("Student/Postdoc")
stu_post_count = data_subset[stu_post == True]
stu_post_df = stu_post_count.groupby('survey_year').agg('count')
stu_post_df['Attendee Type'] = "Student/Professional"

prof = data_all_years_remapped['Ticket'].str.contains("Professional")
prof_count = data_subset[prof == True]
prof_df = prof_count.groupby('survey_year').agg('count')
prof_df['Attendee Type'] = "Professional"

staff = data_all_years_remapped['Ticket'].str.contains("oSTEM Staff")
staff_count = data_subset[staff == True]
staff_df = staff_count.groupby('survey_year').agg('count')
staff_df['Attendee Type'] = "oSTEM Staff"

expo = data_all_years_remapped['Ticket'].str.contains("Expo Only")
expo_count = data_subset[expo == True]
expo_df = expo_count.groupby('survey_year').agg('count')
expo_df['Attendee Type'] = "Expo Only"

frames = [stu_post_df,prof_df,staff_df,expo_df]
df = pd.concat(frames)

df['Year'] = list(df.index)

fig_total_attendee = px.bar(df, x="Year", y="Ticket", color="Attendee Type", 
                 title="Attendee Breakdown at oSTEM Conferences")\
            .for_each_trace(lambda t: t.update(name=t.name.replace("Attendee Type=","")))

fig_total_attendee.layout = {'bargap':0.5, 'barmode':'relative'}

fig_total_attendee.update_layout(
    title="Attendee Breakdown at oSTEM Conferences",
    xaxis_title="Year",
    yaxis_title="Attendee Type",
    xaxis = dict(
        dtick = 1
    ),
    margin={"r":50,"t":100,"l":150,"b":50}
)


#Racial/Ethnic Breakdown

data_subset = data_all_years_remapped[['Racial_Ethnic_Identity_Option1','survey_year']]

white = data_all_years_remapped['Racial_Ethnic_Identity_Option1'].str.contains("White/European")
white_count = white[white == True].shape[0]
asian = data_all_years_remapped['Racial_Ethnic_Identity_Option1'].str.contains("Asian/Pacific Islander")
asian_count = asian[asian == True].shape[0]
hisp = data_all_years_remapped['Racial_Ethnic_Identity_Option1'].str.contains("Chicanx/Latinx/Hispanic")
hisp_count = hisp[hisp == True].shape[0]
black = data_all_years_remapped['Racial_Ethnic_Identity_Option1'].str.contains("Black/African/African American")
black_count = black[black == True].shape[0]
multi = data_all_years_remapped['Racial_Ethnic_Identity_Option1'].str.contains("Multi-Ethnic")
multi_count = multi[multi == True].shape[0]
other = data_all_years_remapped['Racial_Ethnic_Identity_Option1'].str.contains("Other")
other_count = other[other == True].shape[0]

frames = [white_count,asian_count,hisp_count,black_count,multi_count,other_count]

race_ethnicity = pd.DataFrame({'Race/Ethnic Identity': ['White/European','Asian/Pacific Islander',
                                                        'Chicanx/Latinx/Hispanic','Black/African/African American',
                                                        'Multi-Ethnic','Other'],
                               'Number':[white_count,asian_count,hisp_count,black_count,multi_count,other_count]})

fig_race_ethnic = px.pie(race_ethnicity, values='Number', names='Race/Ethnic Identity', 
             title='oSTEM Conference Participants: Race and Ethnic Identity')


fig_race_ethnic.update_layout(
    margin={"r":50,"t":100,"l":100,"b":75}
)

#new comers
First_count1 = data_2020_new['Past_oSTEM_Conferences'].str.contains("oSTEM")
First_count = First_count1[First_count1==True].shape[0]
Returning_count = data_2020_new['Past_oSTEM_Conferences'].str.count(" 2")
Second_count = Returning_count[Returning_count == 0].shape[0] - First_count
Third_count = Returning_count[Returning_count == 1].shape[0]
Fourth_count = Returning_count[Returning_count == 2].shape[0]
Fifth_count = Returning_count[Returning_count == 3].shape[0]
Sixth_count = Returning_count[Returning_count >= 4].shape[0]
Past_Conferences = pd.DataFrame({'Conferences': ["First oSTEM Conference", "Second oSTEM Conference",
                                                 "Third oSTEM Conference", "Fourth oSTEM Conference",
                                                 "Fifth oSTEM Conference", "Sixth or Greater"],
                                 'Count': [First_count, Second_count, Third_count,
                                           Fourth_count, Fifth_count, Sixth_count]})
Returning_pie = px.pie(Past_Conferences, values='Count', names='Conferences',
                       title="Previous Conference Attendance in 2020")

#Sexual Orientation pie chart
gay = data_all_years_remapped['Sexual_Orientation_Option1'].str.contains("Androphilic/Gay")
gay_count = gay[gay == True].shape[0]
bi = data_all_years_remapped['Sexual_Orientation_Option1'].str.contains("Attraction for 2+ genders")
bi_count = bi[bi == True].shape[0]
straight = data_all_years_remapped['Sexual_Orientation_Option1'].str.contains("Straight")
straight_count = straight[straight == True].shape[0]
lesbian = data_all_years_remapped['Sexual_Orientation_Option1'].str.contains("Gynephilic/Lesbian")
lesbian_count = lesbian[lesbian == True].shape[0]
asexual = data_all_years_remapped['Sexual_Orientation_Option1'].str.contains("Asexual/Aromantic Spectrum")
asexual_count = asexual[asexual == True].shape[0]
questioning = data_all_years_remapped['Sexual_Orientation_Option1'].str.contains("Questioning")
questioning_count = questioning[questioning == True].shape[0]
sexual_or_df = pd.DataFrame({'Sexual Orientation': ['Gay','Lesbian','Straight','Bisexual','Asexual','Questioning'],
                             'Number of people': [gay_count,lesbian_count,straight_count,bi_count,asexual_count,questioning_count]})
sexual_or_pie = px.pie(sexual_or_df, values='Number of people', names='Sexual Orientation',
                       title="Sexual Orientation")


#Gender Identity Pie Chart
cis = data_all_years_remapped['Gender_Identity_Option1'].str.contains("Cisgender|neither intersex nor Transgender")
cis_count = cis[cis == True].shape[0]
no_say = data_all_years_remapped['Gender_Identity_Option1'].str.contains("Prefer not to say")
no_say_count = no_say[no_say == True].shape[0]
#The first group 'neither intersex nor transgender' adds people to the transgender field who aren't
transgender = data_all_years_remapped['Gender_Identity_Option1'].str.contains("Transgender")
not_transgender = data_all_years_remapped['Gender_Identity_Option1'].str.contains("neither intersex nor Transgender")
transgender_count = transgender[transgender == True].shape[0] - not_transgender[not_transgender == True].shape[0]
other = data_all_years_remapped['Gender_Identity_Option1'].str.contains("Questioning|Agender/Androgynous|intersex")
other_count = other[other == True].shape[0]
gender_df = pd.DataFrame({'Gender Identity': ['Cisgender','Transgender',
                                              'Prefer not to say','Other'],
                          'Number of people': [cis_count,transgender_count,
                                               no_say_count,other_count]})
gender_pie = px.pie(gender_df, values='Number of people', names='Gender Identity',  
             color_discrete_sequence=px.colors.qualitative.G10, title="Gender Identity")
```
